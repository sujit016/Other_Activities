# Complex Numbers

A **complex number** is a number of the form  

$$
z = a + bi
$$

where:

- $a$ = real part  
- $b$ = imaginary part  
- $i$ = imaginary unit such that  

$$
i^2 = -1
$$


### Basic Components

For $z = a + bi$:

- **Real part:**  
  $$
  \operatorname{Re}(z) = a
  $$

- **Imaginary part:**  
  $$
  \operatorname{Im}(z) = b
  $$


### Conjugate

The conjugate of $z$ is:

$$
\overline{z} = a - bi
$$



### Modulus (Magnitude)

$$
|z| = \sqrt{a^2 + b^2}
$$

This represents the distance from the origin in the complex plane.

### Argument

The argument (angle) of $z$:

$$
\theta = \tan^{-1}\left(\frac{b}{a}\right)
$$


### Polar Form

A complex number can also be written as:

$$
z = r(\cos\theta + i\sin\theta)
$$

where:

- $r = |z|$
- $\theta = \arg(z)$

Using Eulers formula:

$$
z = r e^{i\theta}
$$


### Basic Operations

If  

$$
z_1 = a + bi, \quad z_2 = c + di
$$

- **Addition:**

$$
z_1 + z_2 = (a+c) + (b+d)i
$$

- **Multiplication:**

$$
z_1 z_2 = (ac - bd) + (ad + bc)i
$$

- **Division:**

$$
\frac{z_1}{z_2} = \frac{z_1 \overline{z_2}}{|z_2|^2}
$$


```julia
z1 = 3 + 4im
z2 = 2 - 5im
```

```julia
real(z1)   # real part of complex numbers 
```

```julia
imag(z1)   # imaginary  part of complex numbers
```

```julia
conj(z1)  # conjugate of complex number
```

```julia
abs(z1)   # Modulus (Magnitude)
```

```julia
angle(z1)    # Argument (Angle)  # Angle is returned in radians.
```

```julia
z1 + z2   # addition of complex numbers
```

```julia
z1 - z2   # Subtraction of complex numbers
```

```julia
z1 * z2   # multiplication of complex
```

```julia
z1 / z2 # division of complex numbers
```

## Writing functions in Julia

Functions in Julia are defined using the `function` keyword:

```julia
function add_numbers(a, b)
    return a + b
end

println(add_numbers(3, 5))  
```

Alternatively, use short-form syntax

```julia
add_numbers(a, b) = a + b
println(add_numbers(3, 5))
```

<div style="text-align: justify;">
We can also do customization of the function. Suppose instead of numeric value someone enters a character in the entry, then the summation will not be computed and result in error.
Inside the function body, we can add more comments
</div>

```julia
# Function to determine the nature of roots of a quadratic equation
function quadratic_roots(a, b, c)
    discriminant = b^2 - 4*a*c
    if discriminant > 0
        return "Two distinct real roots"
    elseif discriminant == 0
        return "One real root (repeated)"
    else
        return "Complex conjugate roots"
    end
end

# Test the function
println(quadratic_roots(1, -3, 2))  # Output: Two distinct real roots
println(quadratic_roots(1, -2, 1))  # Output: One real root (repeated)
println(quadratic_roots(1, 2, 5))   # Output: Complex conjugate roots
```

<p style="text-align: justify;">
You can also write down mathematical functions. The common algebraic and trigonometric functions are available in Julia. Some common functions are <code>sin()</code>, <code>cos()</code>, <code>log()</code>, etc. In the following, we show an example. The <code>plot()</code> function from the <code>Plots.jl</code> package is very important to understand when dealing with mathematical functions.

```julia
using Plots, LaTeXStrings
```

```julia
f(x) = x^2*sin(10*x)  # body of the function
p1 = plot(f, -1, 1, color = "red", xlabel = L"x", lw = 3, 
ylabel = L"f(x)", title = L"x^3\sin(10x)", label = "")

g(x) = x^3
h(x) = sin(10*x)
p2 = plot(g, -1, 1, color = "red", lw = 3, linestyle = :dash, 
ylims = (-1.2, 1.6), label = L"g(x)")
plot!(h, color = "blue", lw = 2, linestyle = :dash, xlabel = "x",
    label = L"h(x)")
plot(p1, p2, layout = (1,2), size = (800, 400))
```

```julia
 
x = [1, 10, 4, 2 , 3 , 5]  

println(x)   # Prints the complete vector

println(argmax(x))  # Returns the index of the maximum element in x

println(argmin(x))  # Returns the index of the minimum element in x

println(findall(x .== 1))  # Finds all indices where elements of x are equal to 1

println(findall(x .!= 1))  # Finds all indices where elements of x are NOT equal to 1

println(sum(x .== 1))  # Counts how many times 1 appears in x

println(x[x .== 1])   # Extracts all elements of x that are equal to 1 (logical indexing)


println(x[x .!= 1])  # Extracts all elements of x that are NOT equal to 1

println(x[x .> 1])  # Extracts all elements of x that are greater than 1
```

### Symbolic Computation

<p style="text-align: justify;">
Julia also provides some options for performing symbolic computation using <code>Symbolics.jl</code> package. In the following, we perform the symbolic derivative of the function $f_s(x) = \frac{2x}{3+x}$ to compute \(f_s'(x)\) using Julia. Another example is also given with $g_s(x) = \frac{2x^2}{1+x^2}$ We need to define the function symbolically using the <code>@variables</code> and <code>@derivatives</code> macros before differentiation.
<p>

```julia
using Symbolics, LaTeXStrings
```

```julia
@variables x
f_s = 2*x/(3+x)
```

```julia
df_s = Symbolics.derivative(f_s, x) # derivative w.r.t x
```

```julia
@variables x
g_s =  ((2*x^2)/(1+x^2))
```

```julia
dg_s = Symbolics.derivative(g_s, x)
```

```julia
p1 = plot(df_s, 0.001, 10, color = "red", lw = 3, 
xlabel = L"x" , ylabel = L"\nabla(f_s(x))", label = "")

p2 = plot(dg_s, 0.001, 10, color = "blue", lw = 3, 
xlabel = L"x" , ylabel = L"\nabla(g_s(x))", label = "")

plot(p1, p2, layout = (1,2), size = (800, 500))
```

# Symbolic Integration

Symbolic integration involves finding the exact antiderivative of a function in symbolic form rather than a numerical approximation. In Julia, `SymPy.jl` allows users to define symbolic variables and perform operations such as differentiation and integration efficiently. It provides closed-form solutions whenever possible, making it useful for mathematical analysis, differential equations, and theoretical modeling.

```julia
using SymPy, Plots, LaTeXStrings
x = symbols("x")
expr = sin(x) * exp(x)  # Function to integrate
```

```julia
indef_integral = integrate(expr, x)
```

```julia
p1 = plot(expr, -4, 4, color = "red", lw = 2, xlabel = "x", title = L"e^x\sin(x)", label = "")
p2 = plot(indef_integral, -4, 4, color = "red", lw = 2, xlabel = "x", 
        title = L"\frac {e^x\sin(x)}{2} - \frac {e^x\cos(x)}{2}", label = "")
plot(p1, p2, size = (800, 500), layout = (1, 2))
```
